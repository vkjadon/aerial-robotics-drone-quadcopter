```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns the solution of the equation
def fode(x, t, k):
    solX = - k * x
    return solX
    
# initial condition
x0 = 0.1

# create array of time points
timePoints = np.linspace(0, 10, 11)

print(timePoints)

# solve ODE
k=0.3
solution=odeint(fode, x0, timePoints, args=(k,))
print(solution)

# set data points on plot
plt.plot(timePoints, solution)

# Label X- and Y- axes
plt.xlabel('time')
plt.ylabel('x(t)')

# set the position of text and display desired text
yMax=max(solution)
xMax=max(timePoints)
plt.text(0.8*xMax,0.8*yMax, str("k="+str(k)))

plt.text(0.8*xMax,0.7*yMax, str("x(0)="+str(x0)))

# display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# dy_dt=(u1-y)/tau

# define model
def model(y,t,u):
    k=1.0
    tau=1.0
    dy_dt = (1.0/tau)*(u-y)
    return dy_dt

tf = 60.0                 # final time for simulation
nsteps = 61               # number of time steps
#delta_t = tf/(nsteps-1)   # how long is each time step?
ts = np.linspace(0,tf,nsteps) # linearly spaced time vector


# velocity initial condition
y0 = 10.0
u = np.zeros(nsteps)
u[41:] = 100.0
y = np.zeros(nsteps)

for i in range(1,nsteps):
    # span for next time step
    # solve for next step
    timespan=[ts[i-1],ts[i]]
    sol = odeint(model,y0,timespan,args=(u[i],))
    
    y[i]=sol[0]

# for storing the results


# simulate with ODEINT

    # plot results
plt.subplot(2,1,1)
plt.plot(ts,u,'g:',label='u(t)')
plt.ylabel('input')
plt.xlabel('time')
plt.legend(loc='best')

plt.subplot(2,1,2)
plt.plot(ts,y)
plt.ylabel('response')
plt.xlabel('time')
#plt.legend(loc='best')
plt.show()
```
