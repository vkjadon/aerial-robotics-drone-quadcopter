```python
import numpy as np
import matplotlib.pyplot as plt

tf=600.0                       # final time for simulation
numbers=601                    # number of time ticks

ts=np.linspace(0,tf,numbers)   # linearly spaced time vector

input_signalImpulse=np.zeros(numbers)
input_signalImpulse[200]=1.0

input_signalStep=np.zeros(numbers)
input_signalStep[100:]=1.0

input_signalRamp=0.001*ts

# plot results
plt.plot(ts,input_signalImpulse, 'b-', label='Impulse')
plt.plot(ts,input_signalStep, 'g-', label='Step')
plt.plot(ts,input_signalRamp, 'r--', label='Ramp')

plt.xlabel('Time')
plt.ylabel('Response')
plt.legend(loc='best')
plt.show()
```
